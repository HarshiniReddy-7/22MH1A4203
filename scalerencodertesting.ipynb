```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
#load the data from the csv into a dataframe
df=pd.read_csv('/content/data.csv')
```

```python
#onehotencoding
#perform one-hot encoding on the 'category' column
one_hot_encoder=pd.get_dummies(df['category'])

#label encoding
#initilizing labelencoder
label_encoder=LabelEncoder()
#perform label encoding on the category column
label_encoded=label_encoder.fit_transform(df['category'])
#print the original data
print("original data")
print(df)
#print oneencoder data
print('one hot emcoder')
print(one_hot_encoder)
#print label encoder data
print(label_encoded)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(df[['feature1']])
min_max_scaler = MinMaxScaler()
min_max_scaler_data = min_max_scaler.fit_transform(df[['feature1']])
print('original data')
print(df)
print("standarised data")
print(standardized_data)
print("min max scaler data")
print(min_max_scaler_data)
```

```python
from sklearn.model_selection import train_test_split
#extract features (x) and target variables(y)
#feature are'feature1and 2'
x=df[['feature1','feature2']].values
#target variables is 'target'
y=df['target'].values
#splitting data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
print('training data')
print(x_train)
print('testing data')
print(x_test)
print('ytesting data')
print(y_test)
```
